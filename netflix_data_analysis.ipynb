# Análisis exploratorio de datos de Netflix

Este notebook aplica técnicas de **análisis exploratorio**, **tratamiento de datos** y **visualización** sobre una base de datos de Netflix.

## 1. Librerías y configuración

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='whitegrid')
```

## 2. Carga de datos

```python
from pathlib import Path

candidate_paths = [
    Path('netflix_titles.csv'),
    Path('data/netflix_titles.csv'),
    Path('dataset/netflix_titles.csv'),
]

dataset_path = next((path for path in candidate_paths if path.exists()), None)
if dataset_path is None:
    raise FileNotFoundError('No se encontró netflix_titles.csv en la raíz, data/ o dataset/.')

df = pd.read_csv(dataset_path)
print(f'Dataset cargado desde: {dataset_path}')
df.head()
```

## 3. Exploración inicial

```python
print('Dimensiones del dataset:', df.shape)
df.info()
```

```python
missing_values = df.isna().sum().sort_values(ascending=False)
missing_values[missing_values > 0]
```

## 4. Tratamiento de datos

```python
clean_df = df.copy()

clean_df = clean_df.drop_duplicates()
clean_df['date_added'] = pd.to_datetime(clean_df['date_added'], errors='coerce')
clean_df['country'] = clean_df['country'].fillna('Desconocido')
clean_df['rating'] = clean_df['rating'].fillna('Sin clasificación')
clean_df['duration'] = clean_df['duration'].fillna('No especificada')

clean_df.isna().sum().sort_values(ascending=False).head()
```

## 5. Visualización y análisis

```python
plt.figure(figsize=(7, 4))
sns.countplot(data=clean_df, x='type', hue='type', palette='Set2', legend=False)
plt.title('Distribución por tipo de contenido')
plt.xlabel('Tipo')
plt.ylabel('Cantidad')
plt.tight_layout()
plt.show()
```

```python
top_countries = clean_df['country'].str.split(', ').explode().value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='Blues_d')
plt.title('Top 10 países con más títulos')
plt.xlabel('Número de títulos')
plt.ylabel('País')
plt.tight_layout()
plt.show()
```

```python
year_counts = clean_df['release_year'].value_counts().sort_index()

plt.figure(figsize=(12, 5))
sns.lineplot(x=year_counts.index, y=year_counts.values, color='crimson')
plt.title('Evolución de lanzamientos por año')
plt.xlabel('Año')
plt.ylabel('Cantidad de títulos')
plt.tight_layout()
plt.show()
```

```python
top_genres = clean_df['listed_in'].str.split(', ').explode().value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_genres.values, y=top_genres.index, palette='magma')
plt.title('Top 10 géneros más frecuentes')
plt.xlabel('Número de apariciones')
plt.ylabel('Género')
plt.tight_layout()
plt.show()
```
